```python
import openai
import pandas as pd
import re
import random
```

```python
OPENAI_API_KEY = 'secrete-and-classified'
openai.api_key = OPENAI_API_KEY
```

```python
# Generating 20 ramdon permutation of the lines
def twenty_permutation(size):
    #use a set to avoid duplicate scrambles
    scramble_set = set()
    #only generate 20 scramblings
    while len(scramble_set) < 20:
        #store each scramble in a list
        l = []
        #make sure all lines are in the list
        while len(l) < size:
             val = random.randint(0,size-1)
             #avoid duplicate lines
             if not val in l:
                l.append(val)
        #add the scramble to the set
        #list is not hashable but tuple is (why python? why?)
        scramble_set.add(tuple(l))
        
    return scramble_set
```

```python
#this function query gpt-4 
def gpt4Query(message, temp):
    messages=[{"role": "user", "content": message}]
    response = openai.ChatCompletion.create(
    model="gpt-4",
    max_tokens=1000,
    temperature=temp,
    messages = messages)
    #response is an openai json object, we onlye need the content
    return response["choices"][0]["message"]["content"]
```

```python
#function to read the input file, do all permuations of lines of code for each block of functions in the file
def scramble(file):
    #store all the output in a list
    output_0 = []
    output_07 = []
    f = open(file,'r')
    codes = f.read()
    #use the empty line to separate each block of code
    blank_line_regex = r"(?:\r?\n){2,}"
    codes =  re.split(blank_line_regex, codes.strip())
    #for each block of code
    for i in range(len(codes)):
        #just to know that it is doing something print i
        print("------------ Code: ",i,"------------------")
        lines = codes[i].split('\n')
        setof20 = twenty_permutation(len(lines)-3)

        #for each permuation
        perm = 0
        for tup in setof20:
            print("*******************perm:",perm,"*****************")
            collect_output_0 = []
            collect_output_07 = []
            #create the prompt
            message = lines[1] + "\n" + lines[2] + "\n" 
            for j in list(tup):
                message += lines[j+3] + "\n"
            collect_output_0.append(message)
            collect_output_07.append(message)
            #get the response from gpt-4
            collect_output_0.append(gpt4Query(message, 0))
            collect_output_07.append(gpt4Query(message, 0.7))
            output_0.append(collect_output_0.copy())
            output_07.append(collect_output_07.copy())
            perm += 1
    return output_0, output_07
```

```python
#laissez le beau temps rouler (this may take a while)
#openai server may time out or throw a bad gateway
output0, output07 = scramble('only6.txt')
newdf0 = pd.DataFrame(output0)
newdf0.columns=["querry","response"]
newdf0.to_csv('output_0.csv',index=False)
newdf07 = pd.DataFrame(output07)
newdf07.columns=["querry","response"]
newdf07.to_csv('output_07.csv',index=False)
```
